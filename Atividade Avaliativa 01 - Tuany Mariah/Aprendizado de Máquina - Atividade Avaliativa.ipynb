* g) Extraia outras 5 informações quaisquer de sua base

1) A primeira informação a ser extraída, será a quantidade de atributos que se repetem nas 3 bases de dados fornecidas (KNN, SVM e Naive)

```python
key1 = pd.read_csv('bases/selecionadosKNN.csv')
key2 = pd.read_csv('bases/selecionadosNaive.csv')
key3 = pd.read_csv('bases/selecionadoSVM.csv')
```

```python
# aqui o corre o merge das 3 bases
keybase = pd.merge(key1, key2, how = 'inner',on='class')
keybaseall = pd.merge(keybase, key3, on='class')
keybasedrop = keybaseall.drop('class', axis =1)
```

```python
#aqui ocorre a retirada do _x e _y, feito isso, ocorre a contagem dos atributos que mais se repetem nas 3 bases jun
from collections import Counter
keys = keybasedrop[keybasedrop.columns]

novokeys = []
for x in keys:
    item = x
    for y in ['_x', '_y']:
        item = item.replace(y, "")
    novokeys.append(item)

print('Atributos que mais se repetem nas 3 bases {}'.format(Counter(novokeys)))
```

No Gráfico...

```python
teste = pd.DataFrame.from_dict(Counter(novokeys), orient='index')
teste.plot(kind='bar', legend = False)
```
